```python
#top 10 highest rating iphones on flipkart in india

highest_rated=data.sort_values(by=['Star Rating'],ascending=False)
highest_rated.head(10)
```

```python
#rating of highest rated iphone
```

```python
iphones = highest_rated["Product Name"].value_counts()
label = iphones.index
counts = highest_rated["Number Of Ratings"] # Corrected variable name from 'highest' to 'highest_rated'
figure = px.bar(highest_rated, x=label,
                 y=counts,
                 title="Number of Ratings of Highest Rated iPhones")
figure.show()
```

MOST EXPENSIVE AND LEAST EXPENSIVE

```python
most_expensive = data. loc[data['Sale Price']. idxmax()] # removed newline
least_expensive = data. loc[data['Sale Price']. idxmin()]
# Display the results
print("Most Expensive Product:")
print (most_expensive)
print("\nLeast Expensive Product:")
print(least_expensive)
```
